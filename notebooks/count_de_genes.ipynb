```python
import pandas as pd
```

```python
lfc = 0.5
dfs = [
    f"../data/output/diff_expression_analysis/star_salmon_gene/lfc_{lfc}/MCF10A_WT_vs_MCF10A_T1.tsv.gz",
    f"../data/output/diff_expression_analysis/star_salmon_gene/lfc_{lfc}/MCF10A_T1_vs_MCF10A_C1.tsv.gz",
    f"../data/output/diff_expression_analysis/star_salmon_gene/lfc_{lfc}/MCF10A_WT_vs_MCF10A_C1.tsv.gz",
]
dfs = [pd.read_table(t) for t in dfs]
dfs[0]
```

```python
def print_stats(df, pvalue=0.001):
    dr = df[(df["log2FoldChange"] < 0) & (df["svalue"] < pvalue)]
    nonde = df[df["svalue"] >= pvalue]
    ur = df[(df["log2FoldChange"] > 0) & (df["svalue"] < pvalue)]

    print(len(df), len(nonde), len(dr) + len(ur), len(dr), len(ur))
```

```python
for df in dfs:
    print_stats(df)
```
