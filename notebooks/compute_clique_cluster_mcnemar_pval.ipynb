```python
import pandas as pd
import numpy as np
from statsmodels.stats.contingency_tables import mcnemar
from statsmodels.stats.multitest import fdrcorrection
```

```python
df = pd.read_table(
    "../data/output/comparative_analysis/annotated_domains/10000/clusters/clique.clusters.tsv.gz"
)
df1 = df[df["label"] == "001_MCF10A_REP1"]
df2 = df[df["label"] == "002_MCF10A_REP2"]
df3 = df[df["label"] == "003_MCF10AT1_REP1"]
df4 = df[df["label"] == "004_MCF10AT1_REP2"]
df5 = df[df["label"] == "005_MCF10CA1a_REP1"]
df6 = df[df["label"] == "006_MCF10CA1a_REP2"]

clusters = np.sort(df["cluster"].unique())

df1
```

```python
data = []
for cluster in clusters:
    m = np.zeros([2, 2], dtype=int)
    m[0, 0] = len(df1[df1["cluster"] == cluster])
    m[0, 1] = len(df3[df3["cluster"] == cluster])
    m[1, 0] = len(df2[df2["cluster"] == cluster])
    m[1, 1] = len(df4[df4["cluster"] == cluster])

    res = mcnemar(m, exact=True)
    data.append(["WT", "T1", m, cluster, res.statistic, res.pvalue])

dff1 = pd.DataFrame(
    data, columns=["cond1", "cond2", "m", "cluster", "mcnemar_stat", "mcnemar_pval"]
)
dff1
```

```python
data = []
for cluster in clusters:
    m = np.zeros([2, 2], dtype=int)
    m[0, 0] = len(df1[df1["cluster"] == cluster])
    m[0, 1] = len(df5[df5["cluster"] == cluster])
    m[1, 0] = len(df2[df2["cluster"] == cluster])
    m[1, 1] = len(df6[df6["cluster"] == cluster])

    res = mcnemar(m, exact=True)
    data.append(["WT", "C1", m, cluster, res.statistic, res.pvalue])

dff2 = pd.DataFrame(
    data, columns=["cond1", "cond2", "m", "cluster", "mcnemar_stat", "mcnemar_pval"]
)
dff2
```

```python
df = pd.concat([dff1, dff2])
df["mcnemar_pval_adj"] = fdrcorrection(df["mcnemar_pval"])[1]

df["mcnemar_pval"] = [f"{pv:.2g}" for pv in df["mcnemar_pval"]]
df["mcnemar_pval_adj"] = [f"{pv:.2g}" for pv in df["mcnemar_pval_adj"]]
df.drop(columns=["m"]).to_csv("/tmp/clique_cluster_mcnemar.tsv", sep="\t", index=False)
df
```
